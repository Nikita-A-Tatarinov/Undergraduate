```python
import pandas

data = pandas.read_csv("data.csv")
data.head()
```

```python
data = data.drop(columns=["grade", "provin", "school", "worklocation", "NCEE_science", 
                          "NCEE_art", "nonac", "emptype", "sector"])
data[data['major'] < 3].to_csv("data1_2.csv")
data[data['major'] == 3].to_csv("data3.csv")
data[data['major'] > 3].to_csv("data4_5.csv")
data[data['major'] > 2].sort_values(by=['major']).to_csv("chou_data.csv", index=False)
```

```python
data.loc[data["major"] < 3, ["major"]] = 1
data.loc[data["major"] > 2, ["major"]] = 0
data.to_csv("major_data.csv")
```
